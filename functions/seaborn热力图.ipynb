```python
import numpy as np
import matplotlib.pyplot as plt

atten_token_modes = np.random.rand(512, 1024)

import seaborn as sns
# f,ax = plt.subplots(figsize=(30, 30))
new_x_labels, new_y_labels = [], []
for i in range(atten_token_modes[:,:].shape[1]):
    if i % 6 == 0:
        new_y_labels.append(f'{i-512}')    
    else:
        new_y_labels.append('')
        
        
plt.rcParams['figure.figsize'] = [30, 30]
ax = sns.heatmap(atten_token_modes[:,:].transpose(1, 0), vmax=1, vmin=0, yticklabels=new_y_labels)
```
