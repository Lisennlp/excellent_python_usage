```python
import unicodedata
import sys

for c in range(sys.maxunicode):
    if unicodedata.combining(chr(c)):
        print(chr(c))
        print(c)
```

#### 从上可以看到，sys.maxunicode中包含了Unicode中所有的字符的acssi码，然后通过chr()函数转换为相应字符，在通过unicodedata.combining()查找该字符是否定义，如果没有定义则返回0，即False，否则True。

```python
import unicodedata
import sys
a = 'pýtĥöñ is awesome\n'
cmb_chrs = dict.fromkeys(c for c in range(sys.maxunicode) if unicodedata.combining(chr(c)))
b = unicodedata.normalize('NFD', a)
b
```

#### unicodedata.normalize使用将原始输入标准化为分解形式字符，然后利用构造的cmb_chrs（以Unicode字符为key，value为None）去清洗数据
