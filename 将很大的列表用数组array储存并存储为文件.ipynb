```python
float_datas = array('f', [random() for i in range(10**7)])
print(len(float_datas))
```

```python
write_bin_file = open('float_datas.bin', 'wb')
float_datas.tofile(write_bin_file)
write_bin_file.close()
```

```python
print(float_datas[-1])
print(float_datas[10])
```

```python
# 10000.1
reader_bin_file = open('float_datas.bin', 'rb')
new_float_datas = array('f')
new_float_datas.fromfile(reader_bin_file, 10**7)
reader_bin_file.close()
print(len(new_float_datas))
print(new_float_datas[-1])
print(new_float_datas[10])
```
